```python
import pandas as pd
```

```python
empl_data = pd.read_csv('fixtures/training_data.csv')
```

```python
empl_data.info()
```

```python
empl_data.head()
```

```python
import statsmodels.api as sm
```

```python
y = empl_data.job_growth
```

```python
X = empl_data.ix[:,1:34]
X.head()
```

```python
X = sm.add_constant(X)
```

```python
est = sm.OLS(y,X).fit()
```

```python
est.summary()
```

```python
test_data = pd.read_csv('fixtures/test_data.csv')
```

```python
y_test = test_data.job_growth
X_test = test_data.ix[:,1:34]
```

```python
est.predict(X_test)
```

```python
X_test.head()
```

```python
X_test = sm.add_constant(X_test)
```

```python
y_predicted = est.predict(X_test)
```

```python
print(y_test)
```

```python
SSE = ((y_predicted - y_test)**2).sum()
```

```python
SST = ((y_predicted - y_test.mean())**2).sum()
```

```python
# Sum of Squared Errors/Residuals
print(SSE)
```

```python
# Sum of Squared Total
print(SST)
```

```python
# R-squared
R2 = 1 - (SSE/SST)
print(R2)
```
