```python
import pylab as pl
import numpy as np
```

```python
import statsmodels.api as sm
```

```python
job_growth = empl_data['job_growth_per']
```

```python
empl_level = empl_data['bls_emp_lvl_hsld_thou']
```

```python
fed_funds_rate = empl_data['fed_fund_rate_per']
```

```python
y = np.matrix(job_growth).transpose()
```

```python
x1 = np.matrix(empl_level).transpose()
```

```python
x2 = np.matrix(fed_funds_rate).transpose()
```

```python
weekly_unemp = empl_data['bls_avg_weeks_unemp_per']
```

```python
empl_pop_ratio = empl_data['bls_emp_pop_ratio_per']
```

```python
labor_force_participation = empl_data['bls_lbr_frc_prtcp_per']
```

```python
mortgage_rate = empl_data['30_yr_fixed_mrtg_per']
```

```python
swap_rate = empl_data['swap_rate_per']
```

```python
x1 = np.matrix(empl_level).transpose()
x2 = np.matrix(fed_funds_rate).transpose()
x3 = np.matrix(weekly_unemp).transpose()
x4 = np.matrix(empl_pop_ratio).transpose()
x5 = np.matrix(labor_force_participation).transpose()
x6 = np.matrix(mortgage_rate).transpose()
x7 = np.matrix(swap_rate).transpose()
```

```python
x = np.column_stack([x1,x2,x3,x4,x5,x6,x7])
```

```python
X = sm.add_constant(x)
```

```python
model = sm.OLS(y,X)
f = model.fit()
```
